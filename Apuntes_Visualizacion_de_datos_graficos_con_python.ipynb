```python
# Uso de función para generar gráficos con diferentes paletas de Seaborn
def generar_grafico(palette):

  #Uso de datos del gráfico superior
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y',which='both',size=0)
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False) #Ajustes realizados para la vizualización del gráfico
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')

  # Ajustes anteriormente vistos para mejor vizualización y profesionalismo
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')


  plt.show()
```

```python
# Se trae el código del gráfico y se eliminan los parámetros propios de Matplotlib como los de set_frame, los de visibilidad del eje, dejándolo en falso, y los parámetros de ticks

def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

### Ejercicio extra: Desafío de creación de una animación para comparar diferentes datos


Tu tarea es crear un gráfico animado con Plotly que muestre estos datos. El gráfico debe tener las siguientes características:

1. Dos líneas: una para Brasil y otra para Argentina.
Un botón "Play" para iniciar la animación, mostrando el aumento o disminución del número de inmigrantes a lo largo de los años.
2. Las configuraciones de animación deben hacer que las dos líneas se muestren y animen al mismo tiempo.

```python
# Importación de las bibliotecas
import plotly.graph_objs as go
import pandas as pd

# Creando las variables
anos = list(map(str, range(1980, 2014)))  # Asegurando que 'anos' es una lista de strings

# Extracción de datos de Brasil y Argentina usando el string years
brasil_series = df.loc['Brasil', anos]
argentina_series = df.loc['Argentina', anos]

# Creación de los DataFrames para plotting
datos_brasil = pd.DataFrame({'Año': brasil_series.index.tolist(),
                             'Inmigrantes': brasil_series.values.tolist()})

datos_argentina = pd.DataFrame({'Año': argentina_series.index.tolist(),
                                'Inmigrantes': argentina_series.values.tolist()})

# Creando una figura
fig = go.Figure()

# Agregando la línea con los datos de Brasil y Argentina
fig.add_trace(go.Scatter(x=datos_brasil['Año'], y=datos_brasil['Inmigrantes'], mode='lines', name='Inmigrantes de Brasil', line=dict(width=4)))
fig.add_trace(go.Scatter(x=datos_argentina['Año'], y=datos_argentina['Inmigrantes'], mode='lines', name='Inmigrantes de Argentina', line=dict(width=4)))

# Definiendo las configuraciones de diseño
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Brasil y Argentina a Canadá en el período de 1980 a 2013</b>',
        x=0.1,
        font=dict(size=18)
    ),
    xaxis=dict(title='<b>Año</b>', range=[1980, 2013]),  # Forzar el rango hasta 2013
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200,
    height=600,
    transition={'duration': 100}  # Agregando transición para suavizar la animación
)

# Definiendo las configuraciones de animación
frames = []
for i in range(len(datos_brasil)):
    frame_data = [
        go.Scatter(x=datos_brasil['Año'].iloc[:i+1], y=datos_brasil['Inmigrantes'].iloc[:i+1], mode='lines'),
        go.Scatter(x=datos_argentina['Año'].iloc[:i+1], y=datos_argentina['Inmigrantes'].iloc[:i+1], mode='lines')
    ]
    frames.append(go.Frame(data=frame_data))

# Forzar inclusión del año 2013 en el último frame
frames.append(go.Frame(
    data=[
        go.Scatter(x=datos_brasil['Año'], y=datos_brasil['Inmigrantes'], mode='lines'),
        go.Scatter(x=datos_argentina['Año'], y=datos_argentina['Inmigrantes'], mode='lines')
    ]
))

fig.frames = frames  # Asignando los frames a la figura

# Exportación del gráfico a formato .html
fig.write_html('/content/inmigracion_argentina_y_brasil_a_canada.html')

# Mostrando la figura
fig.show()
```
